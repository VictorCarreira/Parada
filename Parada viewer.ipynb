```python
################PACOTES################
% matplotlib inline
import numpy as np
import pylab as py
import pandas as pd
from pandas import Series, DataFrame
import matplotlib
import matplotlib.pyplot as plt
import scipy as sp
from IPython.display import Latex
#####################################
```

```python
#data = pd.read_csv('Dado.csv')
#data = np.array(data)
data=np.loadtxt('Dado.csv')
print(data)
```

```python

#Gerando o Gráfico de convergência:

x=data[:,0]
y=data[:,1]

py.rcParams['figure.figsize'] = (20.0, 10.0)#Redimensiona a figura

plt.plot( x, y, 'go') # green bolinha
plt.plot( x, y, 'k:', color='orange') # linha pontilha orange

plt.title("$Dado$ $Direto$", fontweight='bold', fontsize=25)
plt.tick_params(axis='both', labelsize=15)#aumenta o tamanho dos valores nos eixos

plt.grid(True)
plt.xlabel("$x$", fontsize=15)
plt.ylabel("$y$", fontsize=15)
plt.savefig('Image/Y.png')  
plt.show()  
plt.close()
```
